# Probability

## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# Total balls
white = 10
red = 12
black = 8
total = white + red + black  # 30 balls

# 1. Probability of taking a white ball
p_white = white / total
print(f"P(white) = {p_white:.2f} or {p_white:.1%}")

# 2. White first, Black second WITHOUT replacement
p_white_then_black_no_replace = (white / total) * (black / (total - 1))
print(f"P(white then black without replacement) = {p_white_then_black_no_replace:.4f} or {p_white_then_black_no_replace:.2%}")

# 3. White first, Black second WITH replacement
p_white_then_black_replace = (white / total) * (black / total)
print(f"P(white then black with replacement) = {p_white_then_black_replace:.4f} or {p_white_then_black_replace:.2%}")

# 4. Three balls, all different colors
# All permutations: W,R,B in any order
from math import comb

# Method: (choose 1 of each color) / (choose 3 from total)
p_all_different = (comb(white, 1) * comb(red, 1) * comb(black, 1)) / comb(total, 3)
print(f"P(all different colors) = {p_all_different:.4f} or {p_all_different:.2%}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# Given:
# P(Cloudy) = 0.4
# P(Rain | Cloudy) = 0.5
# P(Rain | Not Cloudy) = 0.2

# Calculate P(Rain)
p_cloudy = 0.4
p_rain_given_cloudy = 0.5
p_rain_given_not_cloudy = 0.2

p_not_cloudy = 1 - p_cloudy

# Law of total probability
p_rain = (p_rain_given_cloudy * p_cloudy) + (p_rain_given_not_cloudy * p_not_cloudy)

print(f"P(Rain) = {p_rain:.3f} or {p_rain:.1%}")
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
P_A = 0.3
P_B = 0.6
P_A_and_B = 0.1

# Calculate P(A U B) = P(A) + P(B) - P(A ∩ B)
P_A_or_B = P_A + P_B - P_A_and_B

# Check each statement
print(f"Given: P(A)={P_A}, P(B)={P_B}, P(A∩B)={P_A_and_B}")
print(f"P(A U B) = {P_A_or_B}")
print()

statements = [
    ("P(A) = 0.7", 0.7 == P_A),
    ("P(A ∪ B) = 0.6", 0.6 == P_A_or_B),
    ("P(A ∩ B) = 0.2", 0.2 == P_A_and_B),
    ("P(A ∪ A) = 1", 1 == P_A),  # P(A ∪ A) = P(A)
    ("P(A ∩ (B ∪ B)) = 0.4", 0.4 == P_A_and_B),  # B ∪ B = B, so P(A ∩ B)
    ("P(A ∩ B) = 0.3", 0.3 == P_A_and_B),
    ("P(A ∩ B) = 0.9", 0.9 == P_A_and_B)
]

print("TRUE statements:")
for stmt, is_true in statements:
    if is_true:
        print(f"  ✓ {stmt}")
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# P(Covid | China) = 0.01
# P(Covid | Italy) = 0.20
# P(Italy) = 1/6 (rolling a 1)
# P(China) = 5/6 (rolling 2-6)

# We want P(China | Covid)

p_covid_given_china = 0.01
p_covid_given_italy = 0.20
p_china = 5/6
p_italy = 1/6

# Bayes' theorem: P(China|Covid) = P(Covid|China)*P(China) / P(Covid)
# First calculate P(Covid) using law of total probability
p_covid = (p_covid_given_china * p_china) + (p_covid_given_italy * p_italy)

# Now apply Bayes
p_china_given_covid = (p_covid_given_china * p_china) / p_covid

print(f"P(China | Covid) = {p_china_given_covid:.4f} or {p_china_given_covid:.2%}")
```

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Theoretical probabilities
theoretical = {
    7: 6/36,   # (1,6),(2,5),(3,4),(4,3),(5,2),(6,1)
    8: 5/36,   # (2,6),(3,5),(4,4),(5,3),(6,2)
    9: 4/36    # (3,6),(4,5),(5,4),(6,3)
}

print("Theoretical probabilities:")
for num, prob in theoretical.items():
    print(f"P(sum = {num}) = {prob:.4f} or {prob:.2%}")

# Simulation: roll 2 dice 10000 times
n_simulations = 10000
sums = []

for _ in range(n_simulations):
    die1 = np.random.randint(1, 7)
    die2 = np.random.randint(1, 7)
    sums.append(die1 + die2)

# Calculate simulated probabilities
sums_array = np.array(sums)
print("\nSimulated probabilities:")
for num in [7, 8, 9]:
    sim_prob = np.sum(sums_array == num) / n_simulations
    print(f"P(sum = {num}) = {sim_prob:.4f} vs theoretical {theoretical[num]:.4f}")

# Create countplot
plt.figure(figsize=(10, 6))
sns.countplot(x=sums, color='skyblue', edgecolor='black')
plt.title(f'Distribution of Sums from Rolling Two Dice ({n_simulations} simulations)')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Frequency')
plt.axvline(x=6.5, color='red', linestyle='--', alpha=0.5)
plt.axvline(x=7.5, color='red', linestyle='--', alpha=0.5)
plt.axvline(x=8.5, color='red', linestyle='--', alpha=0.5)

# Add theoretical lines for 7,8,9
max_freq = max(np.bincount(sums))
plt.text(7, max_freq * 0.9, f'Sum 7: {theoretical[7]:.1%}', 
         ha='center', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))
plt.text(8, max_freq * 0.8, f'Sum 8: {theoretical[8]:.1%}', 
         ha='center', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))
plt.text(9, max_freq * 0.7, f'Sum 9: {theoretical[9]:.1%}', 
         ha='center', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))

plt.show()

# Bonus: Exact calculation of sums 7,8,9
print("\nDetailed combinations:")
print("Sum = 7: (1,6),(2,5),(3,4),(4,3),(5,2),(6,1) → 6/36 = 1/6 ≈ 16.67%")
print("Sum = 8: (2,6),(3,5),(4,4),(5,3),(6,2) → 5/36 ≈ 13.89%")
print("Sum = 9: (3,6),(4,5),(5,4),(6,3) → 4/36 = 1/9 ≈ 11.11%")
```
